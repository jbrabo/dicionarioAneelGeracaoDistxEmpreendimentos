```python
import tabula

arquivo = "https://dadosabertos.aneel.gov.br/dataset/5e0fafd2-21b9-4d5b-b622-40438d40aba2/resource/3fabb9e8-668a-4f94-8f0e-ed9cd2682979/download/dm-geracao-distribuida-relacao-de-empreendimentos.pdf"

list_tabelas = tabula.read_pdf(arquivo, pages=["3","4","5"])

print(len(list_tabelas))
```

```python
for tabela in list_tabelas:
    display(tabela)
```

```python
df_tab1 = tabula.read_pdf(arquivo, pages=["3"])
df_tab2 = tabula.read_pdf(arquivo, pages=["4"])
df_tab3 = tabula.read_pdf(arquivo, pages=["5"])
```

```python
df_tab1[0].columns = df_tab1[0].iloc[0]
```

```python
df_tab1 = df_tab1[0]
df_tab2 = df_tab2[0]
df_tab3 = df_tab3[0]
```

```python
display(df_tab1)
```

```python
df_tab1 = df_tab1.drop(0)
```

```python
display(df_tab1)
```

```python
df_tab1_corrigida = pd.DataFrame()
df_tab2_corrigida = pd.DataFrame()
df_tab3_corrigida = pd.DataFrame()
```
